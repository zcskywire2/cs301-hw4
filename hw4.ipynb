```python
import os
import numpy as np
import pandas as pd
try:
    df = pd.read_csv("2019_nCoV_20200121_20200206.csv")
except Exception as e:
    print (e.__doc__)
    print (e)
    print("Error during reading, please try again")

def get_percent_nans(df, column_name):
    ret = (df[column_name].isnull().sum()/len(df.index))
    return "{:.2%}".format(ret)

for col in df.columns:
    print("The column "+col+" has "+str(get_percent_nans(df,col))+" missing values")
get_percent_nans(df, 'Death')


df.sort_values(by=['Death'])
print("The Country/Region with the highest number of deaths is: "+ df["Country/Region"][0])


df.sort_values(by=['Suspected'])
print("The Country/Region with the highest number of suspected cases is: "+ df["Country/Region"][0])

#lets do this a stupid way
max = 0
second = 0
print(df[2:3])
for i in range(len(df)):
    percent = df.at[i,'Recovered']/df.at[i,'Confirmed']
    if (percent > max):
        second = max
        max = percent
print ("The second highest percentage of recovered cases is {:.2%}".format(second))

print(df['Country/Region'].unique())
def get_continent(country):
    thisdict = {
        "Mainland China": "Asia",
        'Singapore': "Asia",
        'Thailand': 'Asia',
        'Japan': 'Asia',
        'Hong Kong': 'Asia',
        'South Korea': 'Asia',
        'Taiwan': 'Asia',
        'Germany': 'Europe',
        'Malaysia': 'Asia',
        'Macau': 'Asia',
        'Vietnam': 'Asia',
        'France': 'Europe',
        'United Arab Emirates': 'Asia',
        'Australia': 'Australia',
        'India': 'Asia',
        'Canada': 'North America',
        'Italy': 'Europe',
        'Philippines': 'Asia',
        'Russia': 'Europe',
        'UK': 'Europe',
        'United States': 'North America',
        'Belgium': 'Europe',
        'Cambodia': 'Asia',
        'Finland': 'Europe',
        'Nepal': 'Asia',
        'Spain': 'Europe',
        'Sri Lanka': 'Asia',
        'Sweden': 'Europe',
        'Ivory Coast': 'Africa',
        'Mexico': 'North America',
        'Brazil': 'South America',
        'Colombia': 'South America'    
    }
    return thisdict[country]
df['Continent'] = df['Country/Region'].apply(get_continent)
df.groupby(['Continent']).sum()

```
